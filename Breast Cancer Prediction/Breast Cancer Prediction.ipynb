# Breast Cancer Prediction Using Logistic Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r'C:\Users\qq\Desktop\CERTIFICATIONS\Python\Data Analysis Projects\Breast Cancer Prediction\archive\data.csv')
data.head()
```

```python
data.drop(['id', 'Unnamed: 32'], axis=1, inplace=True)
```

```python
data.head()
```

```python
data.diagnosis = [1 if each == "M" else 0 for each in data.diagnosis]
```

```python
data["diagnosis"] = data["diagnosis"].astype('category',copy=False)
```

```python
data["diagnosis"].value_counts().plot(kind='bar')
```

```python
data.info()
```

Divide into target variable and predictors

```python
y = data["diagnosis"]
x = data.drop("diagnosis", axis=1)
```

```python
x
```

# Normalize the Data

```python
from sklearn.preprocessing import StandardScaler

#Create a Standard Scaler object
scaler = StandardScaler()

#Fit and transform the data
scaler.fit(x)
x_scaled = scaler.transform(x)

#Check the shape
x_scaled
```

# Split The Data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.30, random_state=42)
```

Train the model

```python
from sklearn.linear_model import LogisticRegression

#create a Logistic Regression object
lr = LogisticRegression()

#train the model
lr.fit(X_train, y_train)

#make predictions on target data based on test data
y_pred = lr.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

# Evaluation of Model

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy:{accuracy:.2f}")
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
